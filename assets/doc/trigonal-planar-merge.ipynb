```python
import os
import numpy as np
from angstrom import Molecule
from angstrom.visualize import render, show
from angstrom.visualize.blender import Blender
from IPython.display import Image
```

```python
moldir = "/Users/kbsezginel/kbs/git/lemongrove.github.io/assets/molecules"
mol = Molecule(read=os.path.join(moldir, "intersections", "Trigonal-Planar-n4-m4.xyz"))
```

```python
mol.write("trigonal.pdb", bonds=mol.bonds)
```

```python
mol.get_bonds()
```

```python
mol.bonds
```

```python
mol.center([0, 0, 0])
```

```python
blend = Blender()
blend.config["executable"] = '/Applications/Blender.app/Contents/MacOS/Blender'
blend.config['camera']['type'] = 'PERSP'
blend.config['camera']['zoom'] = 10
blend.config['resolution'] = (1920 / 4, 1080 / 4)
blend.config['use_center'] = False
blend.config['save'] = 'octahedtal.blend'
```

```python
render(mol, renderer=blend)
```
